```python
import pandas as pd 
import numpy as np 
```

```python
df = pd.read_excel('hotelBookings.xlsx')
df
```

```python
''''

we started by deleting any duplicate rows 
57 row was deleted

'''

df = df.drop_duplicates()
```

```python
'''
using describe function we can see if there are any weird values speacially by checking max and min values 

in this data we can see that max number of children is 55000 which menas someone made a reservation for 55000 which is not logical 
same fro adults as 3500 that s also not possible 
arrival data there is one showing 2099 also not logical 

note : ( we don't know what numbers mean in agent and company column maybe should be just deleted ?)

'''


df.describe().loc[['min' , 'max']]
```

```python
df['children'].unique()
```

```python
print(len(df[df['children'] == 10 ]))
print(len(df[df['children'] == 1 ]))
print(len(df[df['children'] == 2 ]))
print(len(df[df['children'] == 0 ]))

'''
we can see that most bookings are with no children with a big diffrence from having 1 , 2 or more children 
so in case of any analyses replacing the exegirated values with 1, 2 or 0 or even deleting the row won't make any difference
in this case we will replace 200 , 55000 with 0 

'''

df['children'][df['children'] > 10 ] = df['children'][df['children'] > 10 ] * 0
```

```python
df['adults'].unique()
```

```python
print(len(df[df['adults'] == 3 ]))
print(len(df[df['adults'] == 1 ]))
print(len(df[df['adults'] == 2 ]))
print(len(df[df['adults'] == 4 ]))
print(len(df[df['adults'] == 3500 ]))

'''
since most adultes are 2 so we will replace that one big value with 2 

'''

df['adults'] = df['adults'].replace(3500 , 2)
```

```python
'''
replace that 2099 date with 2015 

'''

df['arrival_date_year'] = df['arrival_date_year'].replace(2099 , 2015)
```

```python
'''
this way we can see number of empty rows in each column
company column has 932 empty column, crazy !! it s basically useless 

arrival month also has 14 empty row, i guess we can fix that
'''

df.isna().sum()
```

```python
'''
some months were not filled in so wen filter based on month date and week number nd filling in the missig month

'''


df[(df['arrival_date_day_of_month'] <=  31) &( df['arrival_date_week_number'] <= 31)]
```

```python
'''
some months were not filled in so wen filter based on month date and week number nd filling in the missig month

'''

df['arrival_date_month'][(df['arrival_date_day_of_month'] <= 31 ) & (df['arrival_date_week_number'] <= 31 )] = df['arrival_date_month'][(df['arrival_date_day_of_month'] <= 31 ) & (df['arrival_date_week_number'] <= 31 )].fillna('July')
df['arrival_date_month'][ (df['arrival_date_week_number'] >= 31 ) & (df['arrival_date_month'] != 'July')] = df['arrival_date_month'][ (df['arrival_date_week_number'] >= 31 ) & (df['arrival_date_month'] != 'July')].fillna("August")
```

```python
'''
we get ride of spcaes to make our dat more structered 

'''

df['meal'] = df['meal'].str.strip(' ')
```

```python
'''
we can clearly see some numbers ssigned as counties that s not normal 
decision to make : we replce numbers by unknown since we don't know the country name

'''
df['country'].unique()
```

```python
df['country'] = df['country'].apply(lambda x: 'unknown' if  isinstance( x , int) else x )
df['country']  = df['country'].fillna('unknown')
```

```python
df.to_excel('cleaned_hotelbooking.xlsx', index=False)
```
