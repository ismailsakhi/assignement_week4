```python
import pandas as pd 
import numpy as np
```

```python
#start of assignment 1
def create_dataframe(fileName: str) -> pd.DataFrame:
    """Create a dataframe from an Excel file and return it."""
    df = pd.read_excel(fileName)
    return df

df = create_dataframe('hotelBookings.xlsx')
create_dataframe('hotelBookings.xlsx')
```

```python
''''

we started by deleting any duplicate rows 
57 row was deleted

'''
def remove_duplicates(df: pd.DataFrame) -> pd.DataFrame:
    """Remove duplicate rows from the dataframe and return it."""
    df = df.drop_duplicates()
    return df

remove_duplicates(df)
```

```python
'''
using describe function we can see if there are any weird values speacially by checking max and min values 

in this data we can see that max number of children is 55000 which menas someone made a reservation for 55000 which is not logical 
same fro adults as 3500 that s also not possible 
arrival data there is one showing 2099 also not logical 

note : ( we don't know what numbers mean in agent and company column maybe should be just deleted ?)

'''

def check_min_max(df: pd.DataFrame) -> pd.DataFrame:
    """Check the min and max values of the dataframe and return them."""
    return df.describe().loc[['min' , 'max']]

check_min_max(df)
```

```python
def check_unique(df: pd.DataFrame, column: str) -> np.ndarray:
    """Check the unique values in the specified column of the dataframe."""
    return df[column].unique()

check_unique(df, 'children')
```

```python
def check_value_counts(df: pd.DataFrame, column: str) -> pd.Series:
    """Check the value counts of the specified column in the dataframe."""
    return df[column].value_counts()

print(check_value_counts(df, 'children'))

'''
we can see that most bookings are with no children with a big diffrence from having 1 , 2 or more children 
so in case of any analyses replacing the exegirated values with 1, 2 or 0 or even deleting the row won't make any difference
in this case we will replace 200 , 55000 with 0 

'''

df['children'][df['children'] > 10 ] = df['children'][df['children'] > 10 ] * 0
```

```python
'''
replace that 2099 date with 2015 

'''
def replace_date(df: pd.DataFrame, old_year: int, new_year: int) -> pd.DataFrame:
    """Replace old year with new year in the arrival_date_year column of the dataframe."""
    df['arrival_date_year'] = df['arrival_date_year'].replace(old_year , new_year)
    return df

replace_date(df, 2099, 2015)
```

```python
'''
we get rid of spaces to make our data more structered 

'''
def clean_column_spaces(df: pd.DataFrame, column: str) -> pd.DataFrame:
    """Remove leading and trailing spaces from the specified column of the dataframe."""
    df[column] = df[column].str.strip(' ')
    return df

clean_column_spaces(df, 'meal')
```

```python
'''
we can clearly see some numbers assigned as countries that's not normal 
decision to make : we replace numbers by unknown since we don't know the country name

'''
check_unique(df, 'country')
```

```python
def replace_numbers_with_unknown(df: pd.DataFrame, column: str) -> pd.DataFrame:
    """Replace numeric values in the specified column with 'unknown'."""
    df[column] = df[column].apply(lambda x: 'unknown' if  isinstance( x , int) else x )
    df[column]  = df[column].fillna('unknown')
    return df

replace_numbers_with_unknown(df, 'country')
```

```python
def export_dataframe_to_excel(df: pd.DataFrame, file_name: str) -> None:
    """Export the dataframe to an Excel file."""
    df.to_excel(file_name, index=False)

export_dataframe_to_excel(df, 'cleaned_hotelbooking.xlsx')
```

```python
'''
Total sales for each category

'''

def calculate_total_sales(df: pd.DataFrame, column: str) -> pd.DataFrame:
    """Calculate total sales for each in specified column and return a new dataframe."""
    return df.groupby(df[column] , as_index=False)['Sales'].sum(), df2['Sales'].sum()

df_total_sales, total = calculate_total_sales(df2, 'Category')

'''

Percentage that a category contributes to the total sales

'''

def calculate_contribution_percentage(df: pd.DataFrame, total: float) -> pd.DataFrame:
    """Calculate the contribution percentage of each category to the total sales."""
    df['contribution %'] = (df['Sales'] / total) * 100
    return df

df_category_sales = calculate_contribution_percentage(df_total_sales, total)
df_category_sales
```

```python
#start of assignment 3
import pandas as pd
from langdetect import detect, DetectorFactory
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer
from textblob import TextBlob
#For first time use:
nltk.download('vader_lexicon')

# make results deterministic
DetectorFactory.seed = 0  

def detect_language(tweet):
    try:
        return detect(tweet)
    except:
        return 'Unknown'

def analyze_sentiment_english(tweet):
    #Turn tweet into textblob
    testimonial = TextBlob(tweet)
    #Get the polarity score of tweet
    score = testimonial.sentiment.polarity

    #Based on score return if its positive, negative or neutral
    if score > 0:
        return "positive"
    if score < 0:
        return "negative"
    else:
        return "neutral"
    
def analyze_sentiment_other(tweet):
    sid = SentimentIntensityAnalyzer()
    score = sid.polarity_scores(tweet)
    compound_score = score['compound']
    if compound_score >= 0.05:
        return "positive"
    if compound_score <= -0.05:
        return "negative"
    else:
        return "neutral"
    
def analyse_tweets(tweets):
    for index, row in tweets.iterrows():
        if row['Language'] == 'en':
            score = analyze_sentiment_english(str(row['Tweet']))
        else:
            score = analyze_sentiment_other(str(row['Tweet']))
        tweets.loc[index, 'Sentiment'] = score
    return tweets

def main():
    # read tweets.xlsx 
    tweets = create_dataframe("tweets.xlsx")
    tweets['Language'] = tweets['Tweet'].apply(detect_language)
    # save results 
    export_dataframe_to_excel(tweets, "tweets_with_language.xlsx")
    tweets.to_csv("tweets_with_language.csv", index=False)
    print("Done!!! Files saved as: 'tweets_with_language.xlsx' and 'tweets_with_language.csv' ")
    #Add sentiment to tweet
    tweets = analyse_tweets(tweets)
    tweets = tweets[['Tweet', 'Language', 'Sentiment']]
    print(tweets)
    export_dataframe_to_excel(tweets, 'tweets_sentiment.xlsx')
    
main()
```
