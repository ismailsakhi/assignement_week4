```python
import pandas as pd
from langdetect import detect, DetectorFactory

# make results deterministic
DetectorFactory.seed = 0  

# read tweets.xlsx 
df = pd.read_excel("tweets.xlsx")

# find tweet column (default: first if no 'tweet')
tweet_col = None
for c in df.columns:
    if str(c).strip().lower() == "tweet":
        tweet_col = c
        break
if tweet_col is None:
    tweet_col = df.columns[0]

# detect languages 
def safe_detect(text):
    if not isinstance(text, str) or not text.strip():
        return "Unknown"
    try:
        return detect(text)
    except Exception:
        return "Unknown"

df["language"] = [safe_detect(x) for x in df[tweet_col]]

# save results 
df.to_excel("tweets_with_language.xlsx", index=False)
df.to_csv("tweets_with_language.csv", index=False)

print("Done!!! Files saved as: 'tweets_with_language.xlsx' and 'tweets_with_language.csv' ")
df.head()
```
